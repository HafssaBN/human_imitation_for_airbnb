```python
import pandas as pd
df = pd.read_excel(r"C:\Users\hafss\OneDrive\Desktop\airbnbscrapper\Data_out_clickable.xlsx")
df.head()
```

```python
import re

def extract_url(v):
    if isinstance(v, str) and v.startswith('=HYPERLINK('):
        m = re.search(r'"(https?://[^"]+)"', v)
        return m.group(1) if m else v
    return v

for col in ("image", "lien"):
    if col in df.columns:
        df[col] = df[col].map(extract_url)

with pd.option_context("display.max_rows", None, "display.max_columns", None, "display.width", None, "display.max_colwidth", None):
    display(df)
```
